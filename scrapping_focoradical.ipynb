```
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
from google.colab import files as FILE
import os

root_path = os.getcwd()
save_path = os.path.join(os.getcwd(), root_path + "/images/")
```

```
def download(name, url):
  img = requests.get(url).content
  with open('{}{}.jpg'.format(save_path,name), 'wb') as handler:
      handler.write(img)

  #FILE.download('{}{}.jpg'.format(save_path, name))
  #os.remove('{}.jpg'.format(name)) # to save up space
```

```
url = 'https://www.focoradical.com.br/'
request_result = requests.get(url)
soup = BeautifulSoup(request_result.text, "html.parser") 
```

```
race = soup.find_all("div", {"class": "competition-desc"} )
```

```
count = []
for q in range(len(race)):
  c = int(race[q].find("span", {"class": "details-count"}).text.replace(".","").replace(" photos",""))
  count.append(c)

print(sum(count))
```

```
urls = soup.find_all("a", {"class": "competition-wrapper"})
prova = urls[0]['href'].replace("/prova/","")
```

```
raceurl = 'https://www.focoradical.com.br/prova/20191231-virada-carioca-run-2019---rio-de-janeiro'
prova = raceurl.replace("https://www.focoradical.com.br/prova/","")

request_result = requests.get(raceurl)
soup = BeautifulSoup(request_result.text, "html.parser") 
competition_id = soup.find("input", { "id" : "competition_id" })['value']
print(competition_id)
```

```
df = pd.DataFrame()
for runner in range(1, 3000):
  runnerurl = "https://www.focoradical.com.br/busca-numero?type=1&competition_id={}&number={}".format(competition_id, runner)
  request_result = requests.get(runnerurl)
  soup = BeautifulSoup(request_result.text, "html.parser") 
  photos = soup.find_all("img", { "class" : "gallery__photo__thumb__image mobile-custom lazyload" })
  photo = []
  if len(photos) > 0:
    for p in photos:
      photo.append((prova, raceurl, competition_id, runner, p['alt'], p['data-original']))
  #else:
    #print("Não encontrado {}".format(runner))
  
  df = df.append(photo)

df.columns = ['Race','RaceUrl','RaceId','Id', 'Image','ImageUrl']
```
